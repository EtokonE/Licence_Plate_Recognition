```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
history = pd.read_csv('../reports/exp13_7with_aug/history.csv')
```

```python
history.head()
```

```python
def plot_train_val_metric(train_metric: pd.Series, 
                          val_metric: pd.Series, 
                          metric_name: str) -> None:
    plt.figure(figsize=(15, 9))
    plt.plot(train_metric, label=f'train_{metric_name}')
    plt.plot(val_metric, label=f'val_{metric_name}')
    plt.legend(loc='best')
    plt.xlabel('epochs')
    plt.ylabel(f'{metric_name}')
    plt.savefig(f'./{metric_name}.png')
    plt.show()
```

```python
plot_train_val_metric(train_metric=history['train_loss'][2:],
                      val_metric=history['val_loss'][2:],
                      metric_name='loss')
```

```python
plot_train_val_metric(history['train_acc'][2:],
                      history['val_acc'][2:],
                      'Accuracy')
```
